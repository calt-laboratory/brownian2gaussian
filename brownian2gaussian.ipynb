# Transform Brownian noise data into Gaussian data
    1. Create Brownian noise
    2. Rank the Brownian noise
    3. Scale the data
    4. Compute inverse hyperbolic tangent

#### Load modules

```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as stats
```

### 1. Create brownian noise

```python
n = 1000
brownian_noise = np.cumsum(np.random.randn(n)) # apply cumulative sum to normal distributed data
```

```python
plt.plot(brownian_noise, color='red')
plt.title('Brownian Noise')
plt.show()
```

### 2. Rank the brownian noise

```python
# rank data
ranked_data = stats.rankdata(brownian_noise)

# normalize data -> create values between 0 and 1
norm_ranked_data = ranked_data / (n+1)

print(np.min(norm_ranked_data), np.max(norm_ranked_data))
```

### 3. Scale the data

```python
scaled_data = norm_ranked_data * 2 - 1

print(np.min(scaled_data), np.max(scaled_data))
```

### 4. Transform data to Gaussian distribution using inverse hyperbolic tangent

```python
gauss = np.arctanh(scaled_data)
```

### Plot the data

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

ax[0, 0].plot(brownian_noise, color='purple')
ax[0, 0].set_title('Brownian Noise Data')
ax[0, 1].plot(gauss, color='violet')
ax[0, 1].set_title('Transformed (Gaussian) Data')

ax[1, 0].hist(brownian_noise, bins=70, color='purple')
ax[1, 1].hist(gauss, bins=70, color='violet')

plt.show()
```

```python
# brownian2gaussian algorithm wrappend in a function
def brownian2gaussian(n):
    
    # brownian2gaussian algorithm
    brownian_noise = np.cumsum(np.random.randn(n))
    normalized_ranked_data = stats.rankdata(brownian_noise) / (n+1)
    scaled_data = norm_ranked_data * 2 - 1
    gauss = np.arctanh(scaled_data)
    
    # result plots
    fig, ax = plt.subplots(2, 2, figsize=(12, 8))
    ax[0, 0].plot(brownian_noise, color='purple')
    ax[0, 0].set_title('Brownian Noise Data')
    ax[0, 1].plot(gauss, color='violet')
    ax[0, 1].set_title('Transformed (Gaussian) Data')
    ax[1, 0].hist(brownian_noise, bins=70, color='purple')
    ax[1, 1].hist(gauss, bins=70, color='violet')
    plt.show()
```
